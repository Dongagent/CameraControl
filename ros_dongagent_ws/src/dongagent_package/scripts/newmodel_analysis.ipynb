```python
folder_path = "image_analysis/230912Exp14/anger/"
csv_lst = [k for k in os.listdir(folder_path) if '.csv' in k]
# print(csv_lst)
```

```python
import pandas as pd
res_lst = []
root_path = 'image_analysis/230912Exp14/anger/'

for i, v in enumerate(csv_lst):
    a = csv_lst[i]
    tmp = pd.read_csv(os.path.join(root_path, a))
    tmp_res = round(float(tmp['anger']), 7)
    tmp_name = csv_lst[i].replace('_axes_data.csv', '.png')
    res_lst.append([tmp_res, tmp_name])
```
